定义一个随机生成的函数


```python
import pandas as pd
def generate_data():
    """
    随机生成数据
    """
    #规定种子
    np.random.seed(4889)

    x=np.array([10]+list(range(10,29)))
    error=np.round(np.random.randn(20),2)#np.round表示四舍五入一个数组，第一个参数表示数组，第二个参数表示保留的小数位数
    y=x+error
    return pd.DataFrame({"x":x,"y":y})#放入字典，生成一个数据框
```

```python
z=generate_data()
z
```

# 数据可视化

```python
import matplotlib.pyplot as plt #用于画画的
```

```python
def visualize_data(data):
    """
    数据可视化
    """
    fig = plt.figure(figsize=(6,6),dpi=80) #设定图形框大小和分辨率
    ax = fig.add_subplot(111)# 设置只画一幅图，表示画布分成一行一列的第一幅图，这是缩写，参数为(rows=1,cols=1,index=1)
    # 设置标签名和刻度，range中表示每个刻度的位置，从10到30步长为5
    ax.set_xlabel("$x$")
    ax.set_xticks(range(10,31,5))
    ax.set_ylabel("$y$")
    ax.set_yticks(range(10,31,5))
    # 画点图，颜色为蓝色，label是图示
    ax.scatter(data.x,data.y,color="b",label=r"$y=x + \epsilon$")
    plt.legend(shadow=True)# 展示label内容
    plt.show()# 展示图形

    
```

```python
import os
if __name__ == "__main__":
    data = generate_data()
    home_path = os.path.dirname(os.path.abspath(__file__))
    # 粗出数据，win和linux不同
    if os.name == "nt":
        data.to_csv("%s\\simple_example.csv" % home_path, index=False)#win
    else:
        data.to_csv("%s/simple_example.csv" % home_path, index=False)#linux
    visualize_data(data)
```
