```python
#导包，设置
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import matplotlib.pyplot as plt
%matplotlib inline 
#让 matplotlib 绘制的图形直接显示在 Notebook 中，而不是弹出单独的窗口，jupyter独有

import logging #导入 Python 的日志模块，用于控制程序输出信息的级别（调试信息、警告、错误等）
logging.getLogger("tensorflow").setLevel(logging.ERROR) #设置 TensorFlow 的日志级别为 ERROR，屏蔽 TensorFlow 的警告和普通信息，只显示错误信息
"""日志级别（从低到高）
logging.DEBUG     # 最详细：调试信息
logging.INFO      # 一般信息：进度、状态更新
logging.WARNING   # 警告：潜在问题
logging.ERROR     # 错误：出现问题但程序能继续
logging.CRITICAL  # 严重错误：程序可能崩溃
"""
tf.autograph.set_verbosity(0) #设置 TensorFlow 的 AutoGraph 详细程度为 0，关闭 AutoGraph 的调试输出
```

```python
#建立模型
X=np.zeros((1000,400))#X=[n,m]，X是训练集(n,m)，共n个案例和m个输入值
y=np.zeros((1000,1))#y是输出标签(n,1)
model = Sequential(
    [               
        tf.keras.Input(shape=(400,)),    #specify input size，输入向量(m,)
        ### START CODE HERE ### 
        Dense(units=25,activation='sigmoid'),#自行设置该层的神经元数和激活函数
        Dense(units=15,activation='sigmoid'),
        Dense(units=1,activation='sigmoid')
        ### END CODE HERE ### 
    ], name = "my_model" 
)
```

```python
#设置损失函数和训练
model.compile(
    loss=tf.keras.losses.BinaryCrossentropy(),#设置损失函数
    optimizer=tf.keras.optimizers.Adam(0.001),#设置学习率
)

model.fit(#训练模型，X是训练集(n,m)，共n个案例和m个输入值，y是输出标签(n,1)，epochs是训练次数
    X,y,
    epochs=20
)
```

```python
#放入值进行预测，然后进行判断
x=np.zeros((1,400))  #shape为(1,m)
prediction = model.predict(x)
if prediction >= 0.5:
    yhat = 1
else:
    yhat = 0
print(f"prediction after threshold: {yhat}")
```
